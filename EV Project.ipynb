```python
import pandas as pd
import numpy as np
```

```python
#To display all the columns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#Importing the raw data
fuel_data = pd.read_csv('alt_fuel_stations (Nov 7 2022).csv')
```

```python
#Filtering only electric stations using station code 'ELEC'
fuel_data_new = fuel_data[fuel_data['Fuel Type Code'] == 'ELEC']
```

```python
#Replacing NAN with 0
fuel_data_new.fillna(0, inplace = True)
```

```python
fuel_data_new.to_csv('data_analysis.csv')
```

```python
#Filtering to return stations that are in the USA
fuel_data_new1 = fuel_data_new[fuel_data_new['Country'] == 'US']
```

```python
#Adding a new column with 1 for counting numnber of stations
fuel_data_new1['count'] = 1
fuel_data_new1.head()
```

```python
#Removing open date which didnt match other dates
fuel_data_new1 = fuel_data_new1[fuel_data_new1['Open Date'] != '0022-07-22']
```

```python
#Converting to integer
fuel_data_new1['Open Date'] = fuel_data_new1['Open Date'].astype(int)
```

```python
#Profiling the data
fuel_data_new1.head()
```

```python
#Using pivot table to return the newly opened stations per year in each state, company wise
T = pd.pivot_table(fuel_data_new1, index=['Open Date', 'State', 'EV Network'], values = ['count'], aggfunc=np.sum)
T.to_csv('Pivot2.csv')
```

```python
#Removing entries before 2009
Pivi = pd.read_csv('Pivot2.csv')
Pivi = Pivi[Pivi['Open Date']>2008]
Pivi
```

```python
#Using cumsum for cumulative number of stations
Pivi['Grouped Cumulative Sum'] = Pivi[['State','EV Network','count']].groupby(['State','EV Network']).cumsum()
```
